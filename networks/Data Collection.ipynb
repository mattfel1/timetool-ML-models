```python
def get_min_col(df, colname):
    return pd.to_numeric(df[colname], errors="coerce").dropna().min()
```

```python
names = ["keypoints", "depth", "lattices", "loss", "val_loss"]
results = {n : [] for n in names}
for history in histories:
    df = pd.read_csv(history, delimiter=";")
    _, keypoints, _, lattices, depth = history[:-4].split("_")
    results["keypoints"].append(int(keypoints))
    results["depth"].append(int(depth))
    results["lattices"].append(int(lattices))
    results["loss"].append(get_min_col(df, "loss"))
    results["val_loss"].append(get_min_col(df, "val_loss"))
result_df = pd.DataFrame.from_dict(results)
```

```python
print(result_df.to_csv())
```
